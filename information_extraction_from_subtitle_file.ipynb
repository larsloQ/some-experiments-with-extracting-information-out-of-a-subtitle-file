# Information Extraktions aus den Untertiteln der Sendung "SWR Marktcheck vom 08.03.22"

Die URL der Sendung ist: https://www.swrfernsehen.de/marktcheck/sendungen/sendung-vom-8-maerz-2022-104.html
Die Untertitel befinden sich hier: https://api.ardmediathek.de/subtitle-format-service/ebutt/urn:ard:subtitle:6d6b9a6a145df087

Die Themen der Sendung sind:
1. Tipps zum Sparen von Heizkosten
2. Ein Vergleich von Akku-Staubsaugern
3. Mietrecht, Mietminderung bei Schimmel und Lärm erklärt von einem Mietrechts-Experten
4. Vitamin-Zusätze in Lebensmittel 



```python
# Preparation 

import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()

extracted_text_file = os.environ.get('extracted_text_file')
data_folder = os.environ.get('data_folder')
name = extracted_text_file.replace('.txt', '')
```

# a) Named Entity Recognition (NER) 

```python
# Results of Step 4:
print(
    """
    ---
    Organisations found in Data with NER by Flair. (https://github.com/flairNLP/flair)
    using model 'flair/ner-german-large' (Model 1) and 'dbmdz/flair-distilbert-ner-germeval14' (Model 2)
    Details see Step 3
    ---
    """
)

orgas = pd.read_csv(data_folder + '04_' + name + '_ORG.csv').fillna("")
orgas = orgas.rename(columns={'ORG': 'Org. Name', 'count_flair_1': 'Flair-Model 1', 'count_flair_2': 'Flair-Model 2'})
display(orgas)
```

## Ergebnisse Named Entity Recognition (NER) 

- Hersteller und Produkte der Akku-Staubsauger werden extrahiert.
- Verbraucherschutz-Organistationen u.Ä. werden extrahiert
